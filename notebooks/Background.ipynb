# Analyzing Background Documents

## How America Invests 2020 (Vanguard)

[Link](https://personal.vanguard.com/pdf/how-america-invests-2020.pdf)

94% of customers who have at least one taxable account use Vanguard's mutual funds to meet their needs. Of those who only have taxable accounts, 75% use mutual funds.<a id='fn-1-src'></a>[<sup>1</sup>](#fn-1)

Compared to other products that Vanguard offers to retail investors, mutual funds triple the usage of that of individual stocks, including exchange-traded funds.<a id='fn-2-src'></a>[<sup>2</sup>](#fn-2)

About half of the households invested with Vanguard have at least one taxable account. <a id='fn-3-src'></a>[<sup>3</sup>](#fn-3)
